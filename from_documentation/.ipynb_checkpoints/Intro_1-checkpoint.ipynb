```python
print("nodes ", list(G.nodes))
print("edges  ", list(G.edges))
print("Adj for node 1", list(G.adj[1]))
print("neighbors for node spam", list(G.neighbors("spam")))
print("degress for node 1 =", G.degree["spam"])
```

# Drawing graphs

```python
import matplotlib.pyplot as plt
```

```python
G = nx.petersen_graph()
plt.subplot(121);
nx.draw(G, with_labels=True, font_weight='bold')
plt.subplot(122)
nx.draw_shell(G, nlist=[range(6,10), range(6)], with_labels=True, font_weight='bold')
```

```python
# missing Around
G.add_node(100)
G.add_edge(4,100)
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

## Different Shapes of the same NetworkX

```python
G.nodes
```

```python
options = {
    'node_color' : 'green',
    'node_size'  : 400,
    'width'      : 2,
    'with_labels': True,
    'font_weight': 'bold'
}
f = plt.figure(figsize=(10,10))
f.add_subplot(221);
nx.draw_random(G, **options)
f.add_subplot(222);
nx.draw_circular(G, **options)
f.add_subplot(223);
nx.draw_spectral(G, **options)
f.add_subplot(224);
nx.draw_shell(G, nlist=[range(5,10), range(5)], **options)
```

# use multiple shells with draw_shell

```python
G = nx.dodecahedral_graph()
shells = [[2, 3, 4, 5, 6], [8, 1, 0, 19, 18, 17, 16, 15, 14, 7], [9, 10, 11, 12, 13]]
nx.draw_shell(G, nlist=shells, **options)
```
